# Classification on Population and Household Data 

## Answering Classification Question1 Using Logistic Regression - Prediction of the gender distribution of a regional district's population based on the age distribution

```r
library(readxl)
data<-read_excel("proj.xlsx")
```

AGroup1 =  0 to 14
AGroup2 = 15 to 19 
AGroup3 = 20 to 24
AGroup4 = 25 to 64
AGroup5 = 65 to 84
AGroup6 = 85 and above 

```r
data
```

```r
data <- subset(data, !(grepl("^2023", Year) | grepl("^2024", Year) | grepl("^2025", Year) | grepl("^2026", Year) | grepl("^2027", Year) | grepl("^2028", Year)))
```

```r
data
```

```r
data$Sex <- ifelse(data$Sex == "Female", 0, 1)
```

```r
# Fit the logistic regression model
model <- glm(Sex ~ AGroup1 + AGroup2 + AGroup3 + AGroup4 + AGroup5 + AGroup6 , data = data, family = binomial)
```

```r
# Print the model summary
summary(model)
```

## Answering Classification Question2 Using Logistic Regression - Classifing regional districts into different household types based on the average number of persons per household and the age distribution

```r
library(dplyr)
library(tidyr)
library(glmnet)
```

```r
data<-read_excel("proj.xlsx")
```

```r
names(data)[names(data) == "All Ages"] <- "All_Ages"
data
```

```r
household_data<-read_excel("household.xlsx")
```

```r
household_data
```

```r
merged_data <- merge(data, household_data, by = "RD_Code")
```

```r
merged_data
```

```r
# Create a binary variable indicating the household type (1 for a specific type, 0 for others)
merged_data$Year22 <- ifelse(merged_data$Year22 <= 2 & merged_data$AGroup4>= 0 & merged_data$AGroup5 <= 40000, 1, 0)
```

```r

# Select the relevant columns for the logistic regression
logistic_data <- merged_data %>%
  select(District, Year22, AGroup4)
```

```r

# Fit the logistic regression model
model <- glm(Year22 ~ District + AGroup4, data = logistic_data, family = binomial)
```

```r

# Print the model summary
summary(model)
```

## Answering Classification Question3 Using Multinomial Logistic Regression - Prediction of  the regional district of a household based on its demographic characteristics of Age Group. 

```r
data
```

```r
install.packages("nnet")
```

```r
library(nnet)

# Perform multinomial logistic regression
model <- multinom(District ~ AGroup1 + AGroup2 + AGroup3 + AGroup4 + AGroup5 + AGroup6,
              data = data)
```

```r
# Print the summary of the model
summary(model)
```

```r
library(cluster)
```

```r
df <- read.csv ("Population.csv")
names(df)
```

```r
filtered_df <- df[, c('RD.Code','Regional.District','Age.0.14','X15.19','X20.24','X25.64','X65.84','X85.')]
```

```r
normalized_data <- scale(filtered_df[, -c(1, 2)])
```

```r
wss <- sapply(1:10, function(k) {
  kmeans_result <- kmeans(normalized_data, centers = k, nstart = 25)
  sum(kmeans_result$withinss[kmeans_result$size >= k])
})

plot(1:10, wss, type = "b", xlab = "Number of Clusters", ylab = "Within-cluster Sum of Squares")
```

```r
k <- 3  
kmeans_result <- kmeans(normalized_data, centers = k, nstart = 25)
```

```r
cluster_labels <- kmeans_result$cluster
df$Cluster <- cluster_labels
```

```r
aggregate(normalized_data, by = list(Cluster = cluster_labels), FUN = mean)
```

```r
plot(normalized_data[, 1], normalized_data[, 2], col = cluster_labels, pch = 16)
```

```r
selected_columns <- c('Regional.District', 'Sex', 'Year', 'Age.0.14', 'X15.19', 'X20.24', 'X25.64', 'X65.84', 'X85.', 'All.Ages')
df_subset <- df[selected_columns]
```

```r
df_normalized <- scale(df_subset[, 4:10])
```

```r
features <- df_normalized
```

```r
k <- 4  
kmeans_result <- kmeans(features, centers = k)
```

```r
cluster_labels <- kmeans_result$cluster
df_clusters <- cbind(df_subset, Cluster = cluster_labels)
```

```r
print(df_clusters)
```
